```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def plot_julia(c_real, c_imag, max_iter=100):
    width, height = 800, 800
    xmin, xmax = -1.5, 1.5
    ymin, ymax = -1.5, 1.5
    image = np.zeros((height, width))

    c = complex(c_real, c_imag)

    for x in range(width):
        for y in range(height):
            zx = xmin + (x / width) * (xmax - xmin)
            zy = ymin + (y / height) * (ymax - ymin)
            z = complex(zx, zy)
            image[y, x] = julia(z, c, max_iter)

    plt.figure(figsize=(8, 8))
    plt.imshow(image, cmap='twilight', extent=(xmin, xmax, ymin, ymax))
    plt.title(f"Julia para c = {c_real:.3f} + {c_imag:.3f}i")
    plt.show()

def julia(z, c, max_iter):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Interfaz interactiva
interact(
    plot_julia,
    c_real=FloatSlider(value=-0.7, min=-1.5, max=0.5, step=0.01),
    c_imag=FloatSlider(value=0.27, min=-1.0, max=1.0, step=0.01),
    max_iter=(50, 200, 10)
)
```
