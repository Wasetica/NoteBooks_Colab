# LAMBDA

Las funciones lambda en Python son funciones anónimas (sin nombre) que se definen en una sola línea y son útiles para operaciones rápidas y sencillas. Son muy usadas en ciencia de datos con librerías como Pandas o NumPy para aplicar transformaciones a datos.

#ALGUNAS CARACTERÍSTICAS

1.  Anónimas: No tienen nombre (aunque se pueden asignar a variables).
2.  De una sola línea: Solo pueden contener una expresión.
3.  Flexibles: Se usan donde se espera una función (ej: en map(), filter(), sort()).




```python
#Sintaxis básica

lambda argumentos: expresión
```

```python
#Ejemplos sencillos
#Función normal
def suma(a, b):
    return a + b

#Función lambda equivalente
suma_lambda = lambda a, b: a + b

print(suma(3, 4))  # 7
print(suma_lambda(3, 4))  # 7
```

```python
#Uso en map() transformar datos

numeros = [1, 2, 3, 4, 5]
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)  # [1, 4, 9, 16, 25]
```

```python
numeros = [1, 2, 3, 4, 5]
duplicar = lambda x: x * 2
resultado = list(map(duplicar, numeros))
print(resultado)
```

```python
#Uso en filter()(filtrar datos)

pares =lambda x: x % 2 == 0
numeros = [1, 2, 3, 4, 5]
numeros_pares = list(filter(pares, numeros))
print(numeros_pares)  # [2, 4]
```

```python
#Ordenar con sorted()(personalizar criterios)
productos = [('manzana', 1.5), ('banana', 0.8), ('cereza', 2.0)]
productos_ordenados = sorted(productos, key=lambda x: x[1])
print(productos_ordenados)  # [('banana',
```

```python
productos = [("Laptop", 1200), ("Mouse", 25), ("Teclado", 50)]
ordenar_por_precio = lambda producto: producto[1]

print(sorted(productos, key=ordenar_por_precio))
```

```python
#En pandas /Transformar columnas/

import pandas as pd

df = pd.DataFrame({"precios": [10, 20, 30]})
df["precios_con_iva"] = df["precios"].apply(lambda x: x * 1.16)

print(df)
#    precios  precios_con_iva
# 0       10            11.6
# 1       20            23.2
# 2       30            34.8
```

```python
# Código intermedio de lambda

import pandas as pd

# Ejemplo de uso de lambda con Pandas para crear una nueva columna
data = {'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]}
df = pd.DataFrame(data)

# Usando una función lambda para sumar las columnas A y B
df['C'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

# Imprimir el DataFrame resultante
df
```

```python
# Código más avanzado

import pandas as pd

# Crear un DataFrame de ejemplo
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50],
        'C': ['a', 'b', 'c', 'd', 'e']}
df = pd.DataFrame(data)

# Función lambda compleja para aplicar a un DataFrame
# Esta función calcula un valor basado en varias columnas y una condición
complex_lambda = lambda row: (row['A'] * row['B']) if row['C'] in ('a', 'e') else (row['A'] + row['B'])

# Aplicar la función lambda a cada fila del DataFrame
df['D'] = df.apply(complex_lambda, axis=1)


# Función lambda para categorizar valores en otra columna
def categorize_value(value):
    if value < 20:
        return 'Low'
    elif value < 40:
        return 'Medium'
    else:
        return 'High'

# Función lambda para usar la funcion categorize_value en una columna
categorize_lambda = lambda x: categorize_value(x)

df['Category'] = df['D'].apply(categorize_lambda)

# Imprimir el DataFrame resultante
df
```

```python
from matplotlib import pyplot as plt
df['A'].plot(kind='line', figsize=(8, 4), title='A')
plt.gca().spines[['top', 'right']].set_visible(False)
```

#Explicación del código
Tenemos un DataFrame con 3 columnas:


* Columnas numéricas A y B

* Columna categórica C con letras de 'a' a 'e'




#¿Cuándo usar lambda?
Para operaciones simples que no requieran una función definida (def).

Cuando necesitas una función temporal (ej: en argumentos de map(), filter(), sort()).


#Limitaciones


*   No pueden tener bucles o múltiples expresiones.
*   Si la lógica es compleja, es mejor usar def.

#Conclusión

Transformar/filtrar datos.

*   Personalizar operaciones en Pandas/NumPy.
